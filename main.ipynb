```python
import pandas as pd
import numpy as np

# read all csv
appearances_df = pd.read_csv('data/appearances.csv')
games_df = pd.read_csv('data/games.csv')
players_df = pd.read_csv('data/players.csv', encoding='ISO-8859-1')
leagues_df = pd.read_csv('data/leagues.csv')
shots_df = pd.read_csv('data/shots.csv')
teams_df = pd.read_csv('data/teams.csv')
teamstats_df = pd.read_csv('data/teamstats.csv')
```

```python
# ALL TIME SCORES
goals_df = appearances_df.groupby('playerID').sum().sort_values('goals', ascending=False)[['goals']]
top_scorers = players_df.join(goals_df, on='playerID').sort_values('goals', ascending=False)[['playerID', 'name', 'goals']]
top_scorers.head(10)
```

```python
# ASSOCIATION BETWEEN GOALS AND XGOALS
goals_df = appearances_df.groupby('playerID').sum()[['goals', 'xGoals']]
top_scorers = players_df.join(goals_df, on='playerID')[['playerID', 'name', 'goals', 'xGoals']]
top_scorers.sort_values('goals', ascending=False).head(10)
```

```python
# TOP ASSISTS ALL TIME
assists_df = appearances_df.groupby('playerID').sum().sort_values('assists', ascending=False)[['assists']]
top_assists = players_df.join(assists_df, on='playerID').sort_values('assists', ascending=False)[['playerID', 'name', 'assists']]
top_assists.head(10)
```

```python
# TOP KEY PASSER
key_passes_df = appearances_df.groupby('playerID').sum().sort_values('keyPasses', ascending=False)[['keyPasses']]
top_key_passers = players_df.join(key_passes_df, on='playerID').sort_values('keyPasses', ascending=False)[['playerID', 'name', 'keyPasses']]
top_key_passers.head(10)
```

```python
# BEST OFFENSIVE PLAYER
goal_assist_df = appearances_df.groupby('playerID').sum()[['goals', 'assists']]
goal_assist_df['G/A'] = goal_assist_df['goals'] + goal_assist_df['assists']
top_offensive_players = players_df.join(goal_assist_df, on='playerID')[['playerID', 'name', 'G/A']]
top_offensive_players.sort_values('G/A', ascending=False).head(10)
```
